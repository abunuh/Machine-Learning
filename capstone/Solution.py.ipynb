## Reading in the Data

- Unzip the data file `breast-cancer-wisconsin-data`
- Read the data into a Pandas DataFrame

```python
from zipfile import ZipFile
with ZipFile('breast-cancer-wisconsin-data.zip') as zipdata:
    with zipdata.open('data.csv') as rawdata:
        df = pd.read_csv(rawdata)
```

```python
df.head()
```

```python
data = df.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
X = data.drop('diagnosis', axis=1)
y_label = data['diagnosis']
y = y_label.apply(lambda x: int(x=='M')) # 1: Malignant, 0: Benign
```

## Normalization

Based on the boxplots above, we can see that not all features are within the same range.  For a fair comparison and also for the DNN model we will implement later, we will need the feature ranges to be normalized between `[0, 1]`.  We will use the MinMaxScaler preprocessing module to do that.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Xarr = X.to_numpy()
Xn = pd.DataFrame(data=scaler.fit_transform(Xarr), columns=X.columns)
Xn_mean = Xn.iloc[:,0:10]
Xn_se = Xn.iloc[:,10:20]
Xn_worst = Xn.iloc[:,20:30]
```

## Feature Selection

In this section, we will use the Low Variance method to see if there are any features that could be excluded.  We will use a Low Variance threshold value of `0.9 x (1 - 0.9)`.

```python
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=(.9 * (1 - .9)))
Xv = sel.fit_transform(X, y)
```

```python
ndx = sel.get_support(indices=True)
print(f'column indices = {ndx}')
features_rem = np.array2string(X.columns[ndx])
print(f'Feaatures remaining = {features_rem}')
print(f'Number of features remaining = {len(X.columns[ndx])}')
```

```python
Xfs = X.iloc[:,ndx]
Xfs.head()
```

```python
Xnfs = Xn.iloc[:,ndx]
Xnfs.head()
```

## Comment on Low Variance method

The low variance method does not seem to be a good method for selecting features from our data set.  On the normalized feature set, there are no features that can be removed with a 90% low variance threshold.  Since we know that radius, area and perimiter are funtionally related, as well as concavity and concave points, using a correlation method may be more suitable.

## Feature Filtering

Based on the highly correlated features found above, we will drop `area_mean`, `perimeter_mean` and `concave points_mean` from our feature sets.

```python
Xfs = X.drop(['area_mean', 'perimeter_mean', 'concave points_mean'], axis=1)
Xfs.head()
```

```python
Xnfs = Xn.drop(['area_mean', 'perimeter_mean', 'concave points_mean'], axis=1)
Xfs.head()
```

```python
#Generate random seed
import random
random.seed(41)
```

## Split Data into Training and Testing sets

```python
from sklearn.model_selection import train_test_split
Xfs_train, Xfs_test, yfs_train, yfs_test = train_test_split(Xfs, y, test_size=0.2, random_state=19, shuffle=True)
Xnfs_train, Xnfs_test, ynfs_train, ynfs_test = train_test_split(Xnfs, y, test_size=0.2, random_state=114, shuffle=True)
```

## Building the Benchmark Model (AdaBoost)

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import make_classification

t = tm.process_time()

clf = AdaBoostClassifier(n_estimators=1000, random_state=19)
clf.fit(Xfs_train, yfs_train)  
clf.feature_importances_  

yfs_pred = clf.predict(Xfs_test)

elapsed_time = tm.process_time() - t 
```

```python
clf.score(Xfs_train, yfs_train) 
```

```python
feature_importance = dict(zip(Xfs.columns, clf.feature_importances_))
print(feature_importance)
```

```python
feature_importance_df = pd.DataFrame(feature_importance, index=[0])
plt.figure(figsize=(16,8))
plt.xticks(rotation=90)
sns.barplot(data=feature_importance_df)
feature_importance_df.head()
#print(len(Xfs_train.columns), len(clf.feature_importances_))
#print(clf.feature_importances_)
```

## Evaluation Metrics


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy score: ', format(accuracy_score(yfs_test, yfs_pred)))
print('Precision score: ', format(precision_score(yfs_test, yfs_pred)))
print('Recall score: ', format(recall_score(yfs_test, yfs_pred)))
print('F1 score: ', format(f1_score(yfs_test, yfs_pred)))
print('Elapsed Time:', format(elapsed_time))
```

## Tweaking Hyperparameters to improve score

We will change `learning_rate` to `0.8` and `0.9` to try to improve scores.

```python
t = tm.process_time()
clf = AdaBoostClassifier(n_estimators=1000, learning_rate=0.9, random_state=19)
clf.fit(Xfs_train, yfs_train)  
clf.feature_importances_  
yfs_pred = clf.predict(Xfs_test)
elapsed_time = tm.process_time() - t 

print('Accuracy score: ', format(accuracy_score(yfs_test, yfs_pred)))
print('Precision score: ', format(precision_score(yfs_test, yfs_pred)))
print('Recall score: ', format(recall_score(yfs_test, yfs_pred)))
print('F1 score: ', format(f1_score(yfs_test, yfs_pred)))
print('Elapsed Time:', format(elapsed_time))
```

## Building the Deep Neaural Network Model

The DNN model architecture will be the input layer, 2 hidden layers of 16 nodes each and `relu` activation functions and a dense layer for the output with softmax activation function.

```python
from keras.models import Sequential
from keras.layers import Dense

t = tm.process_time()
model = Sequential()
model.add(Dense(16, input_dim=27, activation='relu'))
model.add(Dense(16, input_dim=27, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

## Fit Model

```python
model.fit(Xnfs_train, ynfs_train, epochs=1000, batch_size=32)
```

```python
ynfs_pred = model.predict(Xnfs_test)
elapsed_time = tm.process_time() - t 
```

## Evaluation Metrics


```python
ynfs_pred_df = pd.DataFrame(np.round(ynfs_pred), columns=['Prediction'])
ynfs_pred_df.head()
```

```python
print('Accuracy score: ', format(accuracy_score(ynfs_test, ynfs_pred_df)))
print('Precision score: ', format(precision_score(ynfs_test, ynfs_pred_df)))
print('Recall score: ', format(recall_score(ynfs_test, ynfs_pred_df)))
print('F1 score: ', format(f1_score(ynfs_test, ynfs_pred_df)))
print('Elapsed Time:', format(elapsed_time))
```
